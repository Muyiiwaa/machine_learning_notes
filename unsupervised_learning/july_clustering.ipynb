### CLUSTERING WITH KMEANS 

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# load the dataset

data = load_iris(as_frame=True)['data']
data.head()
```

```python
data.columns
```

```python
sns.set_style('darkgrid')
sns.set_palette('husl')
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))

sns.scatterplot(data=data, x='sepal length (cm)', y='sepal width (cm)', ax=axes[0])
sns.scatterplot(data=data, x='sepal length (cm)', y='petal width (cm)', ax=axes[1])
sns.scatterplot(data=data, x='sepal width (cm)', y='petal length (cm)', ax=axes[2])
```

```python
# build a kmeans model



model = KMeans(n_clusters=4, random_state=23)
model.fit(X=data)
predictions = model.labels_
```

```python
final_data = data.copy()
final_data['clusters'] = [f'cluster_{x}' for x in predictions]
final_data.head(7)
```

```python
sns.set_style('darkgrid')
sns.set_palette('husl')
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

sns.scatterplot(data=final_data, x='sepal length (cm)', y='sepal width (cm)', ax=axes[0], hue='clusters')
sns.scatterplot(data=final_data, x='sepal length (cm)', y='petal width (cm)', ax=axes[1], hue='clusters')
sns.scatterplot(data=final_data, x='sepal width (cm)', y='petal length (cm)', ax=axes[2], hue='clusters')
```

```python
# 3d visual of the clusters


px.scatter_3d(data_frame=final_data, x='sepal length (cm)', y='sepal width (cm)', z ='petal width (cm)',
              color='clusters')
```

### ELBOW CURVE

```python
no_of_clusters = list(range(2, 13))
inertia_list = []

for cluster in no_of_clusters:
    model = KMeans(n_clusters=cluster, random_state=23)
    model.fit(X=data)
    inertia = model.inertia_
    inertia_list.append(inertia)



plt.plot(no_of_clusters, inertia_list, '-o')
plt.title('ELBOW CURVE OF RANGE BETWEEN 3 AND 10')
```
