## RANDOM FORESTS

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.metrics import root_mean_squared_error
```

```python
# load our dataset
X = load_diabetes(as_frame=True)['data']
y = load_diabetes()['target']

X.head()
```

```python
# split the data 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
# train the model

model = RandomForestRegressor(random_state=23, n_estimators=40, max_depth=4)
model.fit(X_train, y_train)
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)
```

```python
print(f'Train rmse:\n {root_mean_squared_error(y_train, train_preds)}')
print(f'test rmse:\n {root_mean_squared_error(y_test, test_preds)}')
```
